```python
#Import Required Dependencies

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings(action="ignore")
```

```python
# Load Data from CSV

loan = pd.read_csv("loan.csv")
loan.head()
```

```python
loan.info()
```

```python
loan.shape
```

```python
loan.isnull().sum()
```

```python
# Removing columns with null values

loan.dropna(axis=1, how='all', inplace=True)
loan.head()
```

```python
loan.shape
```

```python
# Removing columns with more than 90% values as 0

loan.drop(columns=loan.columns[loan.eq(0).mean()>0.9], inplace=True)
loan.head()
```

```python
loan.shape
```

```python
# Some columns after observing have been found to have only single values

loan.nunique()
```

```python
# Will be removing the columns with single type of values

loan.drop(['pymnt_plan', 'initial_list_status', 'policy_code', 'application_type'], axis = 1, inplace=True)
loan.head()
```

```python
# Some of the fields/Columns seem to have no use during the process of loan approval.
# We are dealing with a situation which occurs before a loan is approved
# We will drop those fields

loan.drop(['id', 'member_id', 'url', 'title', 'emp_title', 'zip_code', 'last_credit_pull_d', 'addr_state','desc', 'total_pymnt_inv','funded_amnt', 'delinq_2yrs', 'revol_bal', 'total_pymnt', 'total_rec_prncp', 'total_rec_int', 'recoveries', 'last_pymnt_d', 'last_pymnt_amnt', 'next_pymnt_d', 'mths_since_last_delinq', 'mths_since_last_record'], axis = 1, inplace=True)
loan.head()
```

```python
loan.shape
```

```python
loan.columns.values
```

```python
loan['loan_status'].value_counts(normalize=True)
```

```python
# We see that there are three types of loan status
# We need to find out who are most likely to default loan repayment, and loans with Current status cannot give an insight
# So, will be removing loans records with current status

loan = loan[loan.loan_status != 'Current']
loan['loan_status'].value_counts(normalize=True)
```

```python
# Check for loan data defaulters

sns.countplot(x = 'loan_status', data = loan)
```

```python
charged_off_loan = loan[loan.loan_status == 'Charged Off']
loan.info()
```

```python
# To find columns with missing records

(loan.isna().sum()/len(loan.index))*100
```

```python
loan.emp_length.value_counts()
```

```python
# For emp_length the percentage of missing values is very low, it will be ok to input values

loan.emp_length.fillna(loan.emp_length.mode()[0], inplace = True)
loan.emp_length.isna().sum()
```

```python
# Will be droping rows for revol_util where values are missing

loan.dropna(axis = 0, subset = ['revol_util'] , inplace = True)
loan.revol_util.isna().sum()
```

```python
# 'revol_util' and 'int_rate', needs data to be standardised

loan.revol_util = pd.to_numeric(loan.revol_util.apply(lambda x : x.split('%')[0]))
loan.int_rate = pd.to_numeric(loan.int_rate.apply(lambda x : x.split('%')[0]))
```

```python
# For emp_length we will standardise the data by considering year less than 1 to be 0, and, 10+ to be 10

loan.emp_length = pd.to_numeric(loan.emp_length.apply(lambda x: 0 if '<' in x else (x.split('+')[0] if '+' in x else x.split()[0])))
```
